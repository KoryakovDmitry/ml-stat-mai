```python
from scipy.stats import kstest

column_name = "EmployeeNumber"

data = df[column_name]
a = data.min()
b = data.max()

statistic, p_value = kstest(data, 'uniform', args=(a, b), mode='asymp')

alpha = 0.05
if p_value > alpha:
    print(f"This column: {column_name} looks like uniform")
else:
    print(f"This column: {column_name} doesn't look like uniform")
```

### 2) point

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("datasets/WA_Fn-UseC_-HR-Employee-Attrition.csv")

for column_name in df.columns:
    try:

        median = df[column_name].median()
        mean = df[column_name].mean()
        print("Median:", median)
        print("Mean:", mean)

        plt.hist(df[column_name], bins=20, edgecolor='k')
        plt.axvline(median, color='r', linestyle='dashed', linewidth=1, label='Median')
        plt.axvline(mean, color='g', linestyle='dashed', linewidth=1, label='Mean')

        plt.xlabel(column_name)
        plt.ylabel('Count')
        plt.legend()
        plt.show()

        difference = median - mean
        range_ = max(df[column_name]) - min(df[column_name])
        threshold = 0.15 * range_

        if abs(difference) <= threshold:
            print("Распределение подходит по условию распределения со смещенной медианой относительно среднего.")
        else:
            print("Распределение не подходит по условию распределения со смещенной медианой относительно среднего.")

        print(difference, range_, threshold)
    except:
        pass
```
