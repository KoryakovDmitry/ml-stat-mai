```python
from scipy.stats import kstest

column_name = "EmployeeNumber"

data = df[column_name]
a = data.min()  # минимальное значение в предполагаемом равномерном распределении
b = data.max()  # максимальное значение в предполагаемом равномерном распределении

statistic, p_value = kstest(data, 'uniform', args=(a, b), mode='asymp')

alpha = 0.05
if p_value > alpha:
    print(f"Данные колонки: {column_name} могут быть равномерно распределены")
else:
    print(f"Данные колонки: {column_name} не имеют равномерного распределения")
```

```python
from scipy import stats
import numpy as np

column_name = "Age"

data = df[column_name]
mean = np.mean(data)
std = np.std(data)

# Нормирование данных к стандартному нормальному распределению
z_scores = [(x - mean)/std for x in data]

statistic, p_value = stats.kstest(z_scores, 'norm')

alpha = 0.05
if p_value > alpha:
    print(f"Данные колонки: {column_name} могут быть нормально распределены")
else:
    print(f"Данные колонки: {column_name} не имеют нормальное распределение")
```

```python
from scipy.stats import chi2_contingency
import numpy as np

data = df['x']

observed_frequencies, _ = np.histogram(data, bins='auto')  # получение наблюдаемых частот для каждого интервала

expected_frequencies = np.ones_like(observed_frequencies) * len(data) / len(observed_frequencies)  # ожидаемые частоты

test_statistic, p_value, _, _ = chi2_contingency([observed_frequencies, expected_frequencies])

# p_value - вероятность получить такую же или более экстремальную статистику, если данные равномерно распределены
```
